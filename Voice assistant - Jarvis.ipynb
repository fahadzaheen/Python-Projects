```python
import pyttsx3
import os
import pyjokes
import pyaudio
import speech_recognition as sr
import webbrowser
import datetime
from tkinter import *

def sptext():
    while True:
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            print("listening...")
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)
            try:
                print("Recognizing...")
                data = recognizer.recognize_google(audio)
                print(data)
                return data
            except sr.UnknownValueError:
                print("Not understand")
            
def speecht(x):
    engine = pyttsx3.init()
    voices = engine.getProperty("voices")
    engine.setProperty("voice",voices[1].id)
    rate   = engine.getProperty("rate")
    engine.setProperty("rate",140)
    engine.say(x)
    engine.runAndWait()
    
def voice_assistant():
    while True:

        data1=sptext().lower()

        if "jarvis" in data1 :
            speecht("yes sir how may i help you")

        elif "your name" in data1:
            name = "my name is jarvis"
            speecht(name)

        elif "old are you" in data1:
            age = "i am not a human, but i was built by fahad few minutes ago"
            speecht(age)

        elif "time" in data1:
            time = datetime.datetime.now().strftime("%I%M%p") 
            speecht(time)

        elif "youtube" in data1:
            youtube = webbrowser.open("https://www.youtube.com/")
            ok  = "ok sir"
            speecht(ok)
            speecht(youtube)

        elif "joke" in data1:
            joke = pyjokes.get_joke(language="en",category="all")
            print(joke)
            speecht(joke)

        elif "media" in data1:
            add= "C:/Users/fahadz/Pictures/Camera Roll"
            media = os.listdir(add)
            os.startfile(os.path.join(add,media[6]))
            speecht( "playing media" )

        elif "made" in data1:
            made = "fahad has made me" 
            speecht(made)

        elif "built" in data1:
            built = "i was built by fahad zaheen" 
            speecht(built)

        elif "exit" in data1:
            ok ="ok sir, thank you"
            speecht(ok)
            break

        elif "" in data1:
            note = "did'nt understand, please say again" 
            speecht(note)


sp = Tk()
sp.title(" Jarvis")
sp.geometry("300x450")
sp.config(bg= "green")

lab = Label(sp, text = "Voice assistant", font =('Times New Roman', 20,'bold'),
            bg="green",fg = "white")
lab.place(x=25 , y=50 , height=30 , width=250 )


button = Button(sp, text= "voice", font= ('Times New Roman', 20,'bold'), bg= "orange" ,fg= "blue",
                relief = RAISED, cursor= 'plus', command = voice_assistant)
button.place(x=62.5 ,y=360 ,height=50 ,width=175 )

sp.mainloop()
```
